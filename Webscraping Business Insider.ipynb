# Webscraping Business Insider

```python
import os as os
os.chdir("F:\Python_proj")
import requests
import pandas as pd
from bs4 import BeautifulSoup
from textblob import TextBlob
import datetime
data_f=pd.read_csv("net_data.csv")
date_list=data_f["Date"].tolist()
Senti_Data=[]
Subjective_Data=[]
count=0
for i in date_list:
    date=datetime.datetime.strptime(i, "%m/%d/%Y").strftime("%Y-%m-%d")
    url1 = "http://www.businessinsider.com/archives?date="+date+'&vertical=clusterstock' 
    response1 = requests.get(url1)
    if response1.status_code == 200:
        #print(response.content)
        page_data_soup1 = BeautifulSoup(response1.content,'lxml')
        #print(page_data_soup1.prettify())
        newstable=page_data_soup1.find_all('table')
        a=newstable[0].get_text()
        b=a.split('\n')
        b
        f=list(filter(lambda a: a != '', b))
        f=". ".join(f)
        analysis = TextBlob(f)
        Senti_Data.append(analysis.sentiment.polarity)
        Subjective_Data.append(analysis.sentiment.subjectivity)
        count=count+1
        print(count)
```

```python
Senti_Data=[10*x*(1-y) for x,y in zip(Senti_Data,Subjective_Data)]
```
